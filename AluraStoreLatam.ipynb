### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda2.head()
```

#1. Análisis de facturación



```python
precios_tienda = sum(tienda['Precio'])
print('Tienda1 ->', precios_tienda)

precios_tienda2 = sum(tienda2['Precio'])
print('Tienda2 ->', precios_tienda2)

precios_tienda3 = sum(tienda3['Precio'])
print('Tienda3 ->', precios_tienda3)

precios_tienda4 = sum(tienda4['Precio'])
print('Tienda4 ->', precios_tienda4)
```

# 2. Ventas por categoría

```python
#Tienda 1
cantidad_productos_por_categoria_tienda = tienda.groupby(['Categoría del Producto'])['Producto'].count()
print(cantidad_productos_por_categoria_tienda)
```

```python
#Tienda 2
cantidad_productos_por_categoria_tienda2 = tienda2.groupby(['Categoría del Producto'])['Producto'].count()
print(cantidad_productos_por_categoria_tienda2)
```

```python
#Tienda 3
cantidad_productos_por_categoria_tienda3 = tienda3.groupby(['Categoría del Producto'])['Producto'].count()
print(cantidad_productos_por_categoria_tienda3)
```

```python
#Tienda 4
cantidad_productos_por_categoria_tienda4 = tienda4.groupby(['Categoría del Producto'])['Producto'].count()
print(cantidad_productos_por_categoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
calificacion_clientes_tienda = tienda.groupby(['Producto'])['Calificación'].mean().round(2)
print(calificacion_clientes_tienda)
```

```python
calificacion_clientes_tienda2 = tienda2.groupby(['Producto'])['Calificación'].mean().round(2)
print(calificacion_clientes_tienda2)
```

```python
calificacion_clientes_tienda3 = tienda3.groupby(['Producto'])['Calificación'].mean().round(2)
print(calificacion_clientes_tienda3)
```

```python
calificacion_clientes_tienda4 = tienda4.groupby(['Producto'])['Calificación'].mean().round(2)
print(calificacion_clientes_tienda4)
```

# 4. Productos más y menos vendidos

```python
#Tienda 1
productos_vendidos_tienda1 = tienda['Producto'].value_counts()

producto_mas_vendido = productos_vendidos_tienda1.idxmax()
cantidad_vendida = productos_vendidos_tienda1.max()

print(f'Producto mas vendido: {producto_mas_vendido} ({cantidad_vendida} ventas)')
```

```python
#Tienda
productos_vendidos_tienda2 = tienda2['Producto'].value_counts()

producto_mas_vendido = productos_vendidos_tienda2.idxmax()
cantidad_vendida = productos_vendidos_tienda2.max()

print(f'Producto mas vendido: {producto_mas_vendido} ({cantidad_vendida} ventas)')
```

```python
#Tienda 3
productos_vendidos_tienda3 = tienda3['Producto'].value_counts()

producto_mas_vendido = productos_vendidos_tienda3.idxmax()
cantidad_vendida = productos_vendidos_tienda3.max()

print(f'Producto mas vendido: {producto_mas_vendido} ({cantidad_vendida} ventas)')
```

```python
#Tienda 4
productos_vendidos_tienda4 = tienda4['Producto'].value_counts()

producto_mas_vendido = productos_vendidos_tienda4.idxmax()
cantidad_vendida = productos_vendidos_tienda4.max()

print(f'Producto mas vendido: {producto_mas_vendido} ({cantidad_vendida} ventas)')
```
